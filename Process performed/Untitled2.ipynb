# Guessing Game One

```python
import numpy.random as rdm 

extraction = rdm.randint(1,9)

guess = 0 

c = 0 

while guess != extraction and guess != "exit":
    guess = input("Choose a number between 1 to 9 : ")
    
    if guess.lower() == "exit":
        break
    
    guess = int(guess)
    c += 1
        
    if int(guess) < extraction:
        print("Too low")
    elif int(guess) > extraction:
        print("Too high")
    else:
        print("Right!")
        print("You took only", c, "Tries!")
        continue


```
